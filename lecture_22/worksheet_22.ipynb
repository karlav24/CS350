## Gradient Descent

Recall in Linear Regression we are trying to find the line $$y = X \beta$$ that minimizes the sum of square distances between the predicted `y` and the `y` we observed in our dataset:

$$\mathcal{L}(\mathbf{\beta}) = \Vert \mathbf{y} - X\mathbf{\beta} \Vert^2$$

We were able to find a global minimum to this loss function but we will try to apply gradient descent to find that same solution.

a) Implement the `loss` function to complete the code and plot the loss as a function of beta.

```python
%matplotlib widget
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

beta = np.array([ 1 , .5 ])
xlin = -10.0 + 20.0 * np.random.random(100)
X = np.column_stack([np.ones((len(xlin), 1)), xlin])
y = beta[0]+(beta[1]*xlin)+np.random.randn(100)

fig, ax = plt.subplots()
ax.plot(xlin, y,'ro',markersize=4)
ax.set_xlim(-10, 10)
ax.set_ylim(min(y), max(y))
plt.show()
```

```python
b0 = np.arange(-5, 4, 0.1)
b1 = np.arange(-5, 4, 0.1)
b0, b1 = np.meshgrid(b0, b1)

def loss(X, y, beta):
    # Compute the mean squared error loss
    predictions = X.dot(beta)
    error = predictions - y
    return np.sum(error**2) / (2 * len(y))

def get_cost(B0, B1):
    res = []
    for b0, b1 in zip(B0, B1):
        line = []
        for i in range(len(b0)):
            beta = np.array([b0[i], b1[i]])
            line.append(loss(X, y, beta))
        res.append(line)
    return np.array(res)

cost = get_cost(b0, b1)
 
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
ax.set_xlim(-6, 4)
ax.set_xlabel(r'$\beta_0$')
ax.set_ylabel(r'$\beta_1$')
ax.set_ylim(-5, 4)
ax.set_zlim(0, 30000)

# Creating plot
ax.plot_surface(b0, b1, cost, alpha=.7)
 
# show plot
plt.show()
```

Since the loss is

$$\mathcal{L}(\mathbf{\beta}) = \Vert \mathbf{y} - X\mathbf{\beta} \Vert^2 = \beta^T X^T X \beta - 2\mathbf{\beta}^TX^T\mathbf{y}  + \mathbf{y}^T\mathbf{y}$$

The gradient is

$$\nabla_\beta \mathcal{L}(\mathbf{\beta}) = 2X^T X \beta - 2X^T\mathbf{y}$$

In logistic regression, the `loss` is the negative log-likelihood

$$ \mathcal{l}(\mathbf{\beta}) = - \frac{1}{N} \sum_{i=1}^{N} y_i \log(\sigma(x_i \beta)) + (1 - y_i)\log(1 - \sigma(x_i \beta))$$

the gradient of which is:

$$\nabla_\beta \mathcal{l}(\mathbf{\beta}) = \frac{1}{N} \sum_{i=1}^{N} x_i (y_i - \sigma(x_i \beta)) $$

g) Modify the above code to evaluate the gradient on a random batch of the data. Overlay the true loss curve and the approximation of the loss in your animation.

```python
import numpy as np
from PIL import Image as im
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

TEMPFILE = "temp.png"
true_loss_curve = []
def snap_model(beta, X, y, X_full, y_full, true_loss):
    # Creating figure
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.scatter(X_full[:, 0], X_full[:, 1], color=['g' if label == 1 else 'b' for label in y_full], s=50, alpha=0.5)
    xplot = np.linspace(-5, 5, 50)
    yestplot = - (beta[0] + beta[1] * xplot) / beta[2]  # assuming beta[2] is the coefficient for the intercept term
    ax.plot(xplot, yestplot, 'r-', lw=2, label='Approximation')
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.set_title(f'True Loss: {true_loss:.4f}', fontsize=14)
    ax.legend()
    fig.savefig(TEMPFILE)
    plt.close()
    return im.fromarray(np.asarray(im.open(TEMPFILE)))

def gradient(X_batch, y_batch, beta):
    N = len(y_batch)
    y_pred = sigmoid(X_batch.dot(beta))
    grad = X_batch.T.dot(y_pred - y_batch) / N
    return grad

def sigmoid(x):
    e = np.exp(-x)
    return 1 / (1 + e)

def loss(X, y, beta):
    N = len(y)
    y_pred = sigmoid(X.dot(beta))
    loss = -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
    return loss

def get_cost(B0, B1, X, y):
    res = []
    for b0, b1 in zip(B0, B1):
        line = []
        for i in range(len(b0)):
            beta = np.array([b0[i], b1[i]])
            line.append(loss(X, y, beta))
        res.append(line)
    return np.array(res)

def gradient_descent(X, y, beta_hat, learning_rate, epochs, images, true_loss_curve):
    losses = [loss(X, y, beta_hat)]
    betas = [beta_hat]

    for _ in range(epochs):
        # Random batch selection
        batch_indices = np.random.choice(len(y), size=10, replace=False)
        X_batch, y_batch = X[batch_indices], y[batch_indices]

        # Update beta using the gradient on the batch
        beta_hat = beta_hat - learning_rate * gradient(X_batch, y_batch, beta_hat)

        # Evaluate true loss on the full dataset
        true_loss = loss(X, y, beta_hat)
        true_loss_curve.append(true_loss)

        images.append(snap_model(beta_hat, X_batch, y_batch, X, y, true_loss))
        losses.append(loss(X_batch, y_batch, beta_hat))
        betas.append(beta_hat)
        
    return np.array(betas), np.array(losses)

# Generate data
centers = [[0, 0]]
X_full, y_full = make_blobs(n_samples=100, centers=centers, cluster_std=2, random_state=0)
X = np.column_stack([X_full, np.ones(len(X_full))])  # Add intercept term
y = np.array([1 if label == 1 else 0 for label in y_full])

b0 = np.arange(-5, 5, 0.1)
b1 = np.arange(-5, 5, 0.1)
b0, b1 = np.meshgrid(b0, b1)

beta_start = np.array([0, 0, 0])  # Initial coefficients
learning_rate = 0.1
images = []
betas, losses = gradient_descent(X, y, beta_start, learning_rate, 20, images, [])

images[0].save(
    'logistic_regression_fit_batch.gif',
    optimize=False,
    save_all=True,
    append_images=images[1:],
    loop=0,
    duration=500
)

# Plot the true loss curve and the approximation
plt.figure(figsize=(10, 6))
plt.plot(true_loss_curve, label='True Loss', linewidth=2)
plt.plot(losses, label='Approximation', marker='o', linestyle='dashed')
plt.title('True Loss vs. Approximated Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import numpy as np
from PIL import Image as im
import matplotlib.pyplot as plt

TEMPFILE = "temp.png"

def snap(x, y, pts, losses, grad):
    fig = plt.figure(figsize =(14, 9))
    ax = plt.axes(projection ='3d')
    ax.view_init(20, -20)
    ax.plot_surface(x, y, loss(np.array([x, y])), color='r', alpha=.4)
    ax.plot(np.array(pts)[:,0], np.array(pts)[:,1], losses, 'o-', c='b', markersize=10, zorder=10)
    ax.plot(np.array(pts)[-1,0], np.array(pts)[-1,1], -1, 'o-', c='b', alpha=.5, markersize=7, zorder=10)
    
    # Plot Gradient Vector
    X, Y, Z = [pts[-1][0]], [pts[-1][1]], [-1]
    U, V, W = [-grad[0]], [-grad[1]], [0]
    ax.quiver(X, Y, Z, U, V, W, color='g')
    fig.savefig(TEMPFILE)
    plt.close()
    return im.fromarray(np.asarray(im.open(TEMPFILE)))

def loss(x):
    return np.sin(sum(x**2)) # change this

def gradient(x):
    return 2 * x * np.cos(sum(x**2)) # change this

def gradient_descent(x, y, init, learning_rate, epochs):
    images, losses, pts = [], [loss(init)], [init]
    for _ in range(epochs):
        grad = gradient(init)
        images.append(snap(x, y, pts, losses, grad))
        init = init - learning_rate * grad
        losses.append(loss(init))
        pts.append(init)
    return images

init = np.array([-.5, -.5]) # change this
learning_rate = 1.394 # change this
x, y = np.meshgrid(np.arange(-2, 2, 0.1), np.arange(-2, 2, 0.1)) # change this
images = gradient_descent(x, y, init, learning_rate, 12)

images[0].save(
    'gradient_descent.gif',
    optimize=False,
    save_all=True,
    append_images=images[1:],
    loop=0,
    duration=500
)
```
